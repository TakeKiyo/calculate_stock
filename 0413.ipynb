```python
import pandas as pd
```

```python
df = pd.read_csv("prices_2.csv")
list_A = []
list_B = []
list_C = []
for a in df['Stock A']:
    list_A.append(a);
for b in df['Stock B']:
    list_B.append(b);
for c in df['Stock C']:
    list_C.append(c);
```

```python
# A,B,Cについて、[単位時間ごとの上がり幅,買うべきt,売るべきt,株の種類]をAll_Listに保持
All_List = []
```

```python
# 時間1ごとに上がってたら保持
for t in range(1,64):
    if list_A[t]-list_A[t-1] >0:
        Ran = list_A[t]-list_A[t-1]
        lis = []
        lis.append(Ran)
        lis.append(t-1)
        lis.append(t)
        lis.append("A")
        All_List.append(lis)
for t in range(2,64):
    if list_A[t]-list_A[t-2] >0:
        Ran = (list_A[t]-list_A[t-2])/2
        lis = []
        lis.append(Ran)
        lis.append(t-2)
        lis.append(t)
        lis.append("A")
        All_List.append(lis)
for t in range(3,64):
    if list_A[t]-list_A[t-3] >0:
        Ran = (list_A[t]-list_A[t-3])/3
        lis = []
        lis.append(Ran)
        lis.append(t-3)
        lis.append(t)
        lis.append("A")
        All_List.append(lis)
for t in range(4,64):
    if list_A[t]-list_A[t-4] >0:
        Ran = (list_A[t]-list_A[t-4])/4
        lis = []
        lis.append(Ran)
        lis.append(t-4)
        lis.append(t)
        lis.append("A")
        All_List.append(lis)
```

```python
# 時間1ごとに上がってたら保持
for t in range(1,64):
    if list_B[t]-list_B[t-1] >0:
        Ran = list_B[t]-list_B[t-1]
        lis = []
        lis.append(Ran)
        lis.append(t-1)
        lis.append(t)
        lis.append("B")
        All_List.append(lis)
for t in range(2,64):
    if list_B[t]-list_B[t-2] >0:
        Ran = (list_B[t]-list_B[t-2])/2
        lis = []
        lis.append(Ran)
        lis.append(t-2)
        lis.append(t)
        lis.append("B")
        All_List.append(lis)
for t in range(3,64):
    if list_B[t]-list_B[t-3] >0:
        Ran = (list_B[t]-list_B[t-3])/3
        lis = []
        lis.append(Ran)
        lis.append(t-3)
        lis.append(t)
        lis.append("B")
        All_List.append(lis)
for t in range(4,64):
    if list_B[t]-list_B[t-4] >0:
        Ran = (list_B[t]-list_B[t-4])/4
        lis = []
        lis.append(Ran)
        lis.append(t-4)
        lis.append(t)
        lis.append("B")
        All_List.append(lis)
```

```python
# 時間1ごとに上がってたら保持
for t in range(1,64):
    if list_C[t]-list_C[t-1] >0:
        Ran = list_C[t]-list_C[t-1]
        lis = []
        lis.append(Ran)
        lis.append(t-1)
        lis.append(t)
        lis.append("C")
        All_List.append(lis)
for t in range(2,64):
    if list_C[t]-list_C[t-2] >0:
        Ran = (list_C[t]-list_C[t-2])/2
        lis = []
        lis.append(Ran)
        lis.append(t-2)
        lis.append(t)
        lis.append("C")
        All_List.append(lis)
for t in range(3,64):
    if list_C[t]-list_C[t-3] >0:
        Ran = (list_C[t]-list_C[t-3])/3
        lis = []
        lis.append(Ran)
        lis.append(t-3)
        lis.append(t)
        lis.append("C")
        All_List.append(lis)
for t in range(4,64):
    if list_C[t]-list_C[t-4] >0:
        Ran = (list_C[t]-list_C[t-4])/4
        lis = []
        lis.append(Ran)
        lis.append(t-4)
        lis.append(t)
        lis.append("C")
        All_List.append(lis)
```

```python
# 上がり幅でソート
All_List.sort(reverse=True)
```

```python
# 時間ごとに、どの株を持っているかを決める
flag = []
double_list = []
for i in range(64):
    lis = []
    flag.append(lis)
for item in All_List:
#     print(item)
    ok = True
    # あいだの時間
    if item[2]-item[1] > 1:
        for i in range(item[1]+1,item[2]):
            if len(flag[i]) > 0:
                ok = False
    # 始めの時間
    if len(flag[item[1]]) >= 2:
        ok = False
    # 終わりの時間
    if len(flag[item[2]]) >= 2:
        ok = False
    lis = [item[1],item[2]] 
    if ok:
        if lis not in double_list:
            for j in range(item[1],item[2]+1):
                flag[j].append(item[3])
            double_list.append(lis)
```
