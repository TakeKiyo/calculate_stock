```python
import pandas as pd
```

```python
df = pd.read_csv("prices_2.csv")
list_A = []
list_B = []
list_C = []
for a in df['Stock A']:
    list_A.append(a);
for b in df['Stock B']:
    list_B.append(b);
for c in df['Stock C']:
    list_C.append(c);
```

```python
# A,B,Cについて、[単位時間ごとの上がり幅,買うべきt,売るべきt,株の種類]をAll_Listに保持
All_List = []
```

```python
# 株Aについて 
low_t = 0
low_price = list_A[0]
for t in range(64):
    if t== 0 or list_A[t-1] > list_A[t]:
        if t != 63 and list_A[t] < list_A[t+1]:
            # 下がってたけど上がるところ
            low_t = t
            low_price = list_A[t]
        else:
            # 下がり続けてるところ
            pass

    if t != 0 and list_A[t-1] < list_A[t]:
        if t==63 or list_A[t] > list_A[t+1]:
            # 上がってるけど下がる
            Ran = (list_A[t]-low_price)/(t-low_t)
            lis = []
            lis.append(Ran)
            lis.append(low_t)
            lis.append(t)
            lis.append('A')
            All_List.append(lis)
            # 売る
            if t != 63:
                low_t = t+1
                low_price = list_A[t+1]
        else:
            # 上がり続けるところ
            pass
```

```python
# 上がり幅でソート
All_List.sort(reverse=True)
```

```python
# 時間ごとに、どの株を持っているかを決める
flag = []
double_list = []
for i in range(64):
    lis = []
    flag.append(lis)
for item in All_List:
#     print(item)
    ok = True
    # あいだの時間
    if item[2]-item[1] > 1:
        for i in range(item[1]+1,item[2]):
            if len(flag[i]) > 0:
                ok = False
    # 始めの時間
    if len(flag[item[1]]) >= 2:
        ok = False
    # 終わりの時間
    if len(flag[item[2]]) >= 2:
        ok = False
    lis = [item[1],item[2]] 
    if ok:
        if lis not in double_list:
            for j in range(item[1],item[2]+1):
                flag[j].append(item[3])
            double_list.append(lis)
```
